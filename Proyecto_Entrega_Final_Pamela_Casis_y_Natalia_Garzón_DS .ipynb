```python
#Removemos los Outliers
def remove_outliers(moviesF):
    median = np.median(moviesF)
    std_dev = np.std(moviesF)
    iqr = np.percentile(moviesF, 75) - np.percentile(moviesF, 25)
    threshold = 1 * iqr
    outliers = np.where(np.abs(moviesF - median) > threshold)[0]
    data_limpia = np.delete(moviesF, outliers)
    return data_limpia
```

```python
# Predicción de puntuación de validación cruzada
cv_lasso = regressor_lasso.best_score_

# Predicción de la puntuación R2 de los resultados del conjunto de pruebas
y_pred_lasso_train = regressor_lasso.predict(X_train)
r2_score_lasso_train = r2_score(y_train, y_pred_lasso_train)

# Predicción de la puntuación R2 de los resultados del conjunto de pruebas
y_pred_lasso_test = regressor_lasso.predict(X_test)
r2_score_lasso_test = r2_score(y_test, y_pred_lasso_test)

# Predecir RMSE los resultados del conjunto de prueba
rmse_lasso = (np.sqrt(mean_squared_error(y_test, y_pred_lasso_test)))
print('CV: ', cv_lasso.mean())
print('R2_score (train): ', r2_score_lasso_train)
print('R2_score (test): ', r2_score_lasso_test)
print("RMSE: ", rmse_lasso)
```
